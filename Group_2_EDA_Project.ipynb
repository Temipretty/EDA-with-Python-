```python
#3.How many males and females participated in the games?
merged_data.Sex.value_counts()
```

```python
# 4.What is the total number of medals won?
merged_data['Medal'].count()
```

```python
#5.Which Team has the highest number of medals?
highest_medals = merged_data.groupby('Team')[['Medal']].count().sort_values(by='Medal', ascending=False)
print(highest_medals)
```

```python
 #6.Which Team has the lowest number of medals?
lowest_medals = merged_data.groupby('Team')[['Medal']].count().sort_values(by='Medal', ascending=True)
print(lowest_medals)
```

```python
#7. Which year has the highest medal won?
highest_medalYear = merged_data.groupby('Year')[['Medal']].count().sort_values(by='Medal', ascending=False)
print(highest_medalYear)


```

```python
# 8.Which city participated most over the past 124 years?
MostParticipatedCity = merged_data.groupby('City')[['Year']].value_counts().sort_values(ascending=False)
print(MostParticipatedCity)
```

```python
#9. Which individual player won the highest number of medals?
IndividualPlayerHighestMedal = merged_data.groupby('Name')[['Medal']].value_counts().sort_values(ascending=False)
print(IndividualPlayerHighestMedal)
```

```python
#10. Which place hosted the highest number of games of the past 124 years?
RegionHostedHighest = merged_data.groupby('region')[['Year']].value_counts().sort_values(ascending=False)
print(RegionHostedHighest)
```
